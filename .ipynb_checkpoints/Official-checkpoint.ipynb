## Clean data
Convert `price` of product id 283 (ComBo 49K), row 10 and 11, into `38000.0` as equivalent to row 12

Remove row with `clingme` value in column payment_method

Remove column payment_status due to insufficient and incomplete data

```python
# replace data of row 10, 11 from value 1 to 38000.0
df = df.replace({'price': 1}, {'price': 38000.0})
```

```python
# remove row with value clingme in colunn payment_method
df = df[df['payment_method'] != 'clingme']
df.payment_method.unique()
```

```python
# remove column payment_status
df = df.drop(columns='payment_status')
```

## Pre-process data

Label encoded `state`, `payment_method` as these 2 fields only contain 2 different values: either either `cancel` or `done` and `cod` or `online_payment`

One-hot encode `product_name` to preserve the meaning of categorizing the values

Split `created_date` into `hour` and transform it with sine and cosine formula

Scale `price`, `canceled_history` and `quantity` by using MinMaxScaler

```python
# Set up encoder to encode state, payment_method, product_name
labelEncode_state = preprocessing.LabelEncoder().fit(df.state.values)
labelEncode_payment = preprocessing.LabelEncoder().fit(df.payment_method.values)
```

```python
# Apply label encode to state, payment_method, product_name
df.state = labelEncode_state.transform(df.state)
df.payment_method = labelEncode_payment.transform(df.payment_method)
```

```python
# Convert product_name into numbering label with labelencoder 
# then onehot-encoded it with OneHotEncoder and stored the result as a new dataframe
df.product_name = preprocessing.LabelEncoder().fit_transform(df.product_name)
onehotEncode_product = preprocessing.OneHotEncoder().fit(df.product_name.values.reshape(-1, 1))
product_name_encoded = onehotEncode_product.transform(df.product_name.values.reshape(-1, 1))
product_name_encoded = pd.DataFrame(
                            product_name_encoded.toarray()[:, 1:], 
                            columns=['product{}'.format(i) for i in range(11)])
```

```python
#Concating the onehot-encoded product name back into the original dataframe (df)
df = pd.concat([df.reset_index(drop=True),product_name_encoded.reset_index(drop=True)], axis=1)
```

```python
df.columns
```

```python
# Scale time (hour) with sine and cosine as this is clycic ordinal value
to_date = pd.to_datetime(df.create_date)
df['hour_x'] = np.sin(2*np.pi*(to_date.dt.hour+1)/24)
df['hour_y'] = np.cos(2*np.pi*(to_date.dt.hour+1)/24)
df
```

```python
# Scale price, canceled_history and quantiy via MinMaxScaler
new_df = df.copy()
new_df = new_df.drop(columns=['order_id', 'customer_id', 'product_id', 'create_date', 'state', 'product_name'])
new_df.price = preprocessing.MinMaxScaler().fit_transform(new_df.price.values.reshape(-1, 1))
new_df.canceled_order_history = preprocessing.MinMaxScaler().fit_transform(new_df.canceled_order_history.values.reshape(-1,1))
new_df.quantity = preprocessing.MinMaxScaler().fit_transform(new_df.quantity.values.reshape(-1, 1))
```

```python
new_df
```

## Create model


### Split data into training set and testing test

The ratio of train:test is 0.7

```python
(x_train, x_test, y_train, y_test) = train_test_split(new_df, df.state, train_size=0.7)
# test = new_df.dropna()
# (x_train, x_test, y_train, y_test) = train_test_split(test, df.state, train_size=0.7)
```

### Support Vector Machine

```python
svc = SVC()
y_pred_svc = svc.fit(x_train, y_train).predict(x_test)
result_svc = y_pred_svc == y_test # compare prediction with desirable result
```

```python
'The accuracy of the model {}: {}%'.format(
    'SVC',
    len(result_svc[result_svc])/len(result_svc)*100)
```
